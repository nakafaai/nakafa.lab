# Experiment: Current Curative Evidence Radar

Objective: smoke-test that current evidence records stay separated by route and never become patient-specific action labels.


```python
records = [
    {
        "id": "FDA_CASGEVY",
        "route": "autologous_gene_cell_curative_benchmark",
        "action": "blocked",
    },
    {
        "id": "FDA_ZYNTEGLO",
        "route": "lentiviral_gene_addition_benchmark",
        "action": "blocked",
    },
    {
        "id": "PMID_42274009",
        "route": "pediatric_exa_cel_benchmark",
        "action": "blocked",
    },
    {
        "id": "NCT06465550",
        "route": "asia_autologous_gene_cell_watch",
        "action": "blocked",
    },
    {
        "id": "PMID_41947014",
        "route": "allogeneic_hsct_curative_comparator",
        "action": "blocked",
    },
    {
        "id": "PMID_42261266",
        "route": "disease_modifying_metabolism_comparator",
        "action": "blocked",
    },
]

route_counts = {}
for record in records:
    route_counts[record["route"]] = route_counts.get(record["route"], 0) + 1

assert all(record["action"] == "blocked" for record in records)
assert len(route_counts) == len(records)

decision_summary = {
    "current_state": "curative_evidence_radar_ready_general_only",
    "route_count": len(route_counts),
    "patient_specific_action": "blocked",
}
decision_summary
```

## Result

The radar can support doctor-facing branch questions, but it cannot support patient eligibility, treatment, referral, travel, import, purchase, dosing, or sample-routing decisions.
