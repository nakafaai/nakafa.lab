# Experiment: Case-001 Branch Review Minimum Packet Gate

Objective: smoke-test that branch review stays blocked until all required private packet domains are label-ready.


```python
required_domains = {
    "diagnosis_genotype_phenotype": False,
    "transfusion_burden_response": False,
    "blood_bank_immune_history": False,
    "iron_organ_chelation_status": False,
    "hct_gene_cell_access_context": False,
    "consent_ethics_private_owner_review": False,
}

missing_labels = [
    f"missing_{domain}" for domain, ready in required_domains.items() if not ready
]
packet_ready = all(required_domains.values())

blocked_actions = {
    "eligibility_claim",
    "therapy_selection",
    "trial_screening_instruction",
    "referral_instruction",
    "travel_or_import_plan",
    "purchase_or_procurement_route",
    "dose_or_treatment_recommendation",
    "transfusion_or_chelation_change",
    "sample_routing",
}

assert not packet_ready
assert len(missing_labels) == len(required_domains)
assert "therapy_selection" in blocked_actions

decision_summary = {
    "current_state": "branch_review_packet_not_ready_public_only",
    "missing_domain_count": len(missing_labels),
    "patient_specific_action": "blocked",
}
decision_summary
```

## Result

The public repo can list missing private packet domains. It cannot select a branch or trigger eligibility, referral, trial screening, travel, import, purchase, dosing, transfusion or chelation changes, or sample routing.
