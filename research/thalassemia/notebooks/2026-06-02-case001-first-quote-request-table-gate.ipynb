# Experiment: Case-001 First Quote Request Table Gate

Objective: verify that the June 2 quote table stays public-safe, preclinical, and restricted to the June 1 quote-ready or benchmark items.


```python
from __future__ import annotations

import json
from pathlib import Path

WORKFLOW_PATH = Path(
    "data/workflows/case-001/2026-06-02-first-quote-request-table.json"
)
workflow = json.loads(WORKFLOW_PATH.read_text())
workflow["gate_label"]
```

```python
expected_items = {
    "hydroxyurea",
    "purified resveratrol",
    "sirolimus",
    "melittin",
    "PF-06409577",
}

included_items = {item["item"] for item in workflow["included_quote_items"]}
held_items = {item["item"] for item in workflow["held_out_items"]}
capabilities = set(workflow["requested_lab_capabilities"])
blocked = set(workflow["blocked_public_content"])
decision_rule = workflow["decision_rule"]

assert workflow["gate_label"] == "case001_first_quote_request_table_ready"
assert "preclinical assay quote draft only" in workflow["quote_scope"]
assert included_items == expected_items
assert "standardized 6-shogaol-rich ginger extract" in held_items
assert "T-BDMC-like curcuminoid analog" in held_items
assert any("HbF" in item or "HBG" in item for item in capabilities)
assert any("viability" in item for item in capabilities)
assert any("hemolysis" in item for item in capabilities)
assert "raw data export" in capabilities
assert "patient_samples_without_ethics" in blocked
assert "patient_specific_treatment_selection" in blocked
assert "dose_or_dosing_requests" in blocked
assert "treatment validation" in decision_rule
assert "patient-specific action" in decision_rule

{
    "gate": workflow["gate_label"],
    "included_items": sorted(included_items),
    "held_items": sorted(held_items),
}
```
