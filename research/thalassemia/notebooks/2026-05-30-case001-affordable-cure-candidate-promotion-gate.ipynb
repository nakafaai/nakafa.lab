# Experiment: Case-001 Affordable Cure Candidate Promotion Gate

Objective:
- Verify that May 30 promotion requires gap, benchmark, endpoint, and falsification fields.
- Keep candidate promotion separate from treatment selection.


```python
# Setup: load the public-safe workflow artifact.
from __future__ import annotations

import json
from pathlib import Path

DATA_PATH = Path(
    "data/workflows/case-001/2026-05-30-affordable-cure-candidate-promotion-gate.json"
)

with DATA_PATH.open() as handle:
    workflow = json.load(handle)

workflow["gate_label"]
```

```python
required_fields = {
    "candidate_category",
    "target_gap_axis",
    "closest_curative_benchmark",
    "closest_noncurative_benchmark",
    "endpoint_set",
    "safety_stop_rule",
    "access_claim_status",
    "falsification_test",
    "case_relevance_label",
}

required_decisions = {
    "promote_to_gap_review",
    "hold_for_endpoint_or_identity_gap",
    "benchmark_only",
    "reject_hazard_or_unmeasured_claim",
    "case_relevance_blocked_until_specialist_review",
}

blocked_content = {
    "patient_specific_treatment_selection",
    "trial_screening_instruction",
    "travel_or_import_instruction",
}

decisions = {item["decision"] for item in workflow["promotion_decision_rules"]}
lane_states = {item["promotion_state"] for item in workflow["current_lane_labels"]}

assert (
    workflow["gate_label"] == "case001_affordable_cure_candidate_promotion_gate_ready"
)
assert set(workflow["required_promotion_fields"]) == required_fields
assert decisions == required_decisions
assert blocked_content <= set(workflow["blocked_public_content"])
assert "benchmark_only" in lane_states
assert "reject_hazard_or_unmeasured_claim" in lane_states

summary = {
    "required_field_count": len(workflow["required_promotion_fields"]),
    "decision_count": len(workflow["promotion_decision_rules"]),
    "lane_count": len(workflow["current_lane_labels"]),
}
summary
```

## Results

- The workflow requires nine promotion fields.
- The workflow supports promote, hold, benchmark-only, reject, and case-blocked decisions.
- The workflow blocks treatment selection, trial screening, and travel or import instructions.
