# Experiment: Case-001 Mitapivat Boundary Access Inquiry Gate

Objective: separate adult label, adult TDT, pediatric TDT, and pediatric NTDT mitapivat contexts before clinician inquiry.

```python
from __future__ import annotations

import json
from pathlib import Path

registry_paths = {
    "adult_tdt_trial_context": Path(
        "data/registries/clinicaltrials/2026-05-11-nct04770779-energize-t-detail.json"
    ),
    "pediatric_tdt_trial_context": Path(
        "data/registries/clinicaltrials/2026-05-11-nct07506863-mitapivat-pediatric-tdt-detail.json"
    ),
    "pediatric_ntdt_trial_context": Path(
        "data/registries/clinicaltrials/2026-05-11-nct07517133-mitapivat-pediatric-ntdt-detail.json"
    ),
}


def first_record(path: Path) -> dict[str, object]:
    """Return the first compact ClinicalTrials.gov record from a snapshot."""
    payload = json.loads(path.read_text())
    records = payload.get("records", [])
    if not records:
        raise ValueError(f"No registry records in {path}")
    return records[0]


registry_contexts = {
    context: first_record(path) for context, path in registry_paths.items()
}

allowed_contexts = {
    "approved_adult_label_context",
    "adult_tdt_trial_context",
    "pediatric_tdt_trial_context",
    "pediatric_ntdt_trial_context",
}

blocked_claims = [
    "patient_specific_eligibility",
    "dosing_or_start_stop_advice",
    "adult_label_applies_to_pediatric_case",
    "registry_presence_means_local_access",
    "raw_record_public_release",
]

decision = {
    "gate_label": "mitapivat_boundary_access_inquiry_ready",
    "allowed_contexts": sorted(allowed_contexts),
    "registry_contexts": registry_contexts,
    "blocked_claims": blocked_claims,
    "policy": "ask_for_context_review_not_treatment_decision",
}

decision
```

## Decision

Continue with mitapivat boundary and access inquiry. Adult label evidence and pediatric registry context stay separate until qualified owner review.
