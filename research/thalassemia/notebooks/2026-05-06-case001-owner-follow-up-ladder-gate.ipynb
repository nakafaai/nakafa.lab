# Experiment: Case-001 Owner Follow-Up Ladder Gate

Objective: map owner outreach states to narrow next actions without raw communication or patient-specific treatment claims.


```python
follow_up_ladder = {
    "not_sent": ("send_owner_row_only", "follow_up_not_due"),
    "ready_to_follow_up": ("prepare_one_narrow_domain_question", "follow_up_ready"),
    "sent_no_reply": ("ask_if_this_is_the_right_owner", "follow_up_sent"),
    "records_requested": ("return_to_record_request_matrix", "records_pending"),
    "redirected": ("return_to_specialist_owner_routing", "owner_redirected"),
    "cannot_review": (
        "choose_alternate_owner_or_leave_unresolved",
        "alternate_owner_needed",
    ),
    "reply_with_content": ("use_owner_response_capture", "response_captured_private"),
    "label_closed": ("release_checked_label_update_only", "route_closed_label_only"),
    "identifier_or_treatment_instruction": (
        "keep_private_and_block_release",
        "release_blocked",
    ),
}

decision = {
    "gate_label": "owner_follow_up_ladder_ready",
    "allowed_input_states": sorted(follow_up_ladder),
    "allowed_public_states": sorted({state for _, state in follow_up_ladder.values()}),
    "default_public_policy": "state_only_after_release_check",
    "patient_specific_claims": "blocked",
}

decision
```
