# Experiment: Case-001 Asia Curative Trial Geography Gate

Objective:
- Validate the May 25 public registry map for curative gene-cell therapy geography.
- Keep the result as benchmark and owner-routing evidence, not treatment advice.

Success criteria:
- Indonesia active thalassemia registry count is recorded as zero for this pass.
- China has at least one curated curative registry signal.
- CS-101, YOLT-204, lentiviral, and exa-cel context remain separated.
- Blocked public content is explicit.


```python
from __future__ import annotations

import json
from pathlib import Path

DATA_PATH = Path(
    "../../../data/registries/clinicaltrials/"
    "2026-05-25-asia-curative-gene-cell-therapy-refresh.json"
)

payload = json.loads(DATA_PATH.read_text(encoding="utf-8"))
payload["evidence_label"]
```

## Plan

- Check the date, evidence label, and status boundary.
- Check country-count invariants for Indonesia and China.
- Check the curated NCT set contains the high-value benchmark lanes.
- Check the public-release blockers prevent trial screening and access advice.


```python
assert payload["checked_date"] == "2026-05-25"
assert payload["evidence_label"] == "asia_curative_gene_cell_therapy_registry_refresh"
assert "not treatment advice" in payload["status"]

counts = payload["active_thalassemia_country_counts"]
assert counts["Indonesia"] == 0
assert counts["China"] > counts["Malaysia"]
assert counts["China"] >= 30

counts
```

```python
records = payload["curated_registry_signals"]
records_by_id = {record["nct_id"]: record for record in records}

required_ids = {
    "NCT07489196",  # CS-101 phase 2
    "NCT06479616",  # CS-101 follow-up
    "NCT07190001",  # YOLT-204
    "NCT05762510",  # LentiRed
    "NCT06280378",  # KL003
    "NCT06308159",  # vebeglogene autotemcel
    "NCT05477563",  # CTX001 regional context
}

missing_ids = required_ids - records_by_id.keys()
assert not missing_ids, missing_ids

assert records_by_id["NCT07489196"]["phase"] == ["PHASE2"]
assert records_by_id["NCT07489196"]["status"] == "NOT_YET_RECRUITING"
assert "China" in records_by_id["NCT07190001"]["country_signal"]

sorted(records_by_id)
```

```python
labels = set(payload["operational_labels"])
assert "asia_curative_gene_cell_therapy_refresh_ready" in labels
assert "indonesia_active_thalassemia_registry_evidence_absent_current_pass" in labels
assert "case001_advanced_therapy_referral_packet_missing" in labels
assert "specialist_screening_required" in labels

blocked = set(payload["blocked_public_content"])
assert "trial enrollment instructions" in blocked
assert "eligibility screening for a person" in blocked
assert "travel or import advice" in blocked
assert "treatment choice or dosing" in blocked

payload["decision_rule"]
```

## Results

- The registry gate is internally consistent.
- The China signal is a benchmark cluster, not case access.
- The Indonesia result is an evidence-absence label for this pass, not proof that no pathway can ever exist.
- Case-001 stays blocked at `advanced_therapy_referral_packet_missing` until a specialist-readable packet is complete.


## Next steps

- Keep this as a benchmark geography watchlist.
- Do not publish trial-screening, travel, import, or patient-eligibility instructions.
- Use the doctor-facing question in the case-context guide before any referral-center discussion.
