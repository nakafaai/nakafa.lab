# Experiment: Case-001 Curative Branch Doctor Question Gate

Objective: smoke-test that every branch output stays a question label and never becomes a patient-specific action.


```python
allowed_labels = {
    "packet_not_ready",
    "branch_question_ready_general_only",
    "standard_care_stabilization_first",
    "allogeneic_hct_branch_owner_question",
    "autologous_gene_cell_branch_owner_question",
    "disease_modifying_noncurative_branch_owner_question",
    "benchmark_only_tracking",
    "route_owner_question_only",
}

blocked_actions = {
    "eligibility_claim",
    "trial_screening_instruction",
    "referral_instruction",
    "travel_or_import_plan",
    "purchase_or_procurement_route",
    "dose_or_treatment_recommendation",
    "transfusion_or_chelation_change",
    "sample_routing",
}

candidate_outputs = [
    "packet_not_ready",
    "branch_question_ready_general_only",
    "allogeneic_hct_branch_owner_question",
    "autologous_gene_cell_branch_owner_question",
    "disease_modifying_noncurative_branch_owner_question",
    "benchmark_only_tracking",
    "route_owner_question_only",
]

assert set(candidate_outputs).issubset(allowed_labels)
assert not set(candidate_outputs).intersection(blocked_actions)

decision_summary = {
    "current_state": "curative_branch_doctor_question_ready_general_only",
    "allowed_output_count": len(candidate_outputs),
    "patient_specific_action": "blocked",
}
decision_summary
```

## Result

The gate can support a clinician-facing branch question, but it cannot support eligibility, referral, trial screening, travel, import, purchase, dosing, transfusion or chelation changes, or sample routing.
