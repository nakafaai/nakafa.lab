# Experiment: Case-001 T-BDMC Material Identity Gate

Objective: verify that `T-BDMC` has exact database identity but remains held until procurement, batch, and safety-window requirements pass.


```python
from __future__ import annotations

import json
from pathlib import Path

WORKFLOW_PATH = Path(
    "data/workflows/case-001/2026-06-04-t-bdmc-material-identity-gate.json"
)
workflow = json.loads(WORKFLOW_PATH.read_text())
workflow["gate_label"]
```

```python
identity = workflow["resolved_identity"]
requirements = set(workflow["material_requirements_before_quote"])
blocked = set(workflow["blocked_public_content"])

assert workflow["gate_label"] == "case001_t_bdmc_material_identity_gate_hold"
assert identity["pubchem_cid"] == 10447050
assert identity["chembl_id"] == "CHEMBL469419"
assert identity["inchikey"] == "PALMCMYYFAHUGA-BPTNNVFMSA-N"
assert (
    workflow["quote_state"]
    == "exclude_from_first_quote_until_material_requirements_pass"
)
assert any("certificate of analysis" in item for item in requirements)
assert any("no substitution" in item for item in requirements)
assert any("hemolysis" in item for item in requirements)
assert "patient_samples_without_ethics" in blocked
assert "dose_or_dosing_requests" in blocked
assert "patient_specific_treatment_selection" in blocked

{
    "gate": workflow["gate_label"],
    "quote_state": workflow["quote_state"],
    "inchikey": identity["inchikey"],
}
```
