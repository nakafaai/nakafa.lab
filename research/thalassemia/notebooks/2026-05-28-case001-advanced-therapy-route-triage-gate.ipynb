# Case-001 Advanced-Therapy Route Triage Gate

This notebook checks that advanced-therapy route triage separates branch
questions without selecting therapy, inferring access, or publishing private
HLA, donor, trial, travel, import, or dosing content.

```python
routes = {
    "allogeneic_hsct_branch_owner_question": "transplant_branch",
    "autologous_gene_cell_branch_owner_question": "gene_cell_branch",
    "disease_modifying_noncurative_branch_owner_question": "drug_branch",
    "standard_care_stabilization_first": "stabilization_branch",
    "trial_registry_benchmark_only": "registry_branch",
}

blocked_public_content = {
    "raw_records",
    "doctor_messages",
    "owner_replies",
    "exact_private_dates",
    "hla_or_donor_details",
    "family_identifiers",
    "product_access_replies",
    "patient_specific_treatment_selection",
    "trial_screening_instruction",
    "travel_or_import_instruction",
    "dose_or_dosing_requests",
}

allowed_public_labels = set(routes) | {
    "case001_advanced_therapy_route_triage_ready",
    "route_selection_blocked_until_specialist_review",
    "release_blocked_private_material",
}

errors = []
if len(routes) != 5:
    errors.append("route branch count changed")
if "allogeneic_hsct_branch_owner_question" not in allowed_public_labels:
    errors.append("HSCT branch label is missing")
if "autologous_gene_cell_branch_owner_question" not in allowed_public_labels:
    errors.append("gene-cell branch label is missing")
if "hla_or_donor_details" not in blocked_public_content:
    errors.append("HLA or donor details must remain blocked")
if "patient_specific_treatment_selection" not in blocked_public_content:
    errors.append("treatment selection must remain blocked")
if "trial_screening_instruction" not in blocked_public_content:
    errors.append("trial-screening instruction must remain blocked")

assert not errors, errors

{
    "route_count": len(routes),
    "allowed_public_label_count": len(allowed_public_labels),
    "blocked_public_content_count": len(blocked_public_content),
}
```
