# Experiment: AND017 Oral Benchmark Gate

Objective: decide whether `NCT06302491` / AND017 changes the closest non-editing benchmark for Nakafa Lab.

Success criteria:
- load the redacted compact ClinicalTrials.gov snapshot;
- score endpoint overlap against Nakafa's non-editing gates;
- keep the result as benchmark logic, not therapy advice.


```python
from __future__ import annotations

import json
from pathlib import Path

SNAPSHOT = (
    Path.cwd()
    / "data/registries/clinicaltrials"
    / "2026-04-28-nct06302491-and017-detail.json"
)

record = json.loads(SNAPSHOT.read_text(encoding="utf-8"))["records"][0]
record
```

## Plan

Hypothesis: AND017 is a useful oral non-editing benchmark for hemoglobin, transfusion-load, and iron-marker endpoints, but it does not displace `NCT07338344` as the closest non-editing pharmacologic proximity benchmark because this compact record does not include HbF, gamma-globin, hemolysis, alpha-globin, or autophagy endpoints.


```python
endpoint_text = " ".join(
    [*record["primary_outcomes"], *record["secondary_outcome_groups"]]
).lower()

gate_terms = {
    "safety": ["safety", "tolerability"],
    "hemoglobin_or_rbc": ["hemoglobin", "red blood cell", "reticulocyte"],
    "transfusion_burden": ["transfusion load", "transfusion count"],
    "iron_markers": ["ferritin", "serum iron", "tsat", "tibc"],
    "hbf_or_gamma_globin": ["hbf", "gamma-globin", "gamma globin"],
    "hemolysis": ["hemolysis"],
    "alpha_globin_autophagy": ["alpha-globin", "autophagy"],
}

coverage = {
    name: any(term in endpoint_text for term in terms)
    for name, terms in gate_terms.items()
}

coverage
```

## Results

`NCT06302491` is important because it is oral, randomized, placebo-controlled, phase 2, recruiting, China-located, and explicitly covers TDT and NTDT cohorts. It adds a practical non-editing comparator for hemoglobin, transfusion load, red-cell indices, and iron markers.

It should not be promoted as an HbF rescue benchmark from this source snapshot. The compact record supports an oral anemia/transfusion benchmark role, not a cure claim and not an integrated HbF plus alpha-globin/autophagy benchmark.


```python
decision = {
    "record": record["nct_id"],
    "role": "oral_non_editing_anemia_transfusion_benchmark",
    "does_not_displace": (
        "NCT07338344 for HbF/gamma-globin and hemolysis-rich pharmacologic proximity"
    ),
    "promote_as": (
        "watchlist comparator for Hb, transfusion-load, red-cell-index, "
        "and iron-marker endpoints"
    ),
    "do_not_claim": "HbF rescue, alpha-globin cleanup, cure, or patient suitability",
}

decision
```
