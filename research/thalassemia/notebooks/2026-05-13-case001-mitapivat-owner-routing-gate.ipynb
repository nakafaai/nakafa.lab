# Experiment: Case-001 Mitapivat Owner Routing Gate

Objective: route each mitapivat access or relevance question to one qualified owner before outreach.

```python
from __future__ import annotations

import json
from pathlib import Path

route_map = json.loads(
    Path(
        "data/regulatory/mitapivat/2026-05-13-mitapivat-owner-route-map.json"
    ).read_text()
)

owners = {record["owner"]: record for record in route_map["owner_routes"]}
safe_states = {record["safe_state"] for record in route_map["owner_routes"]}
required_owners = {
    "treating_hematologist",
    "hospital_pharmacy_or_specialty_pharmacy",
    "regulator_or_regulatory_affairs_contact",
    "manufacturer_medical_information",
    "trial_site_or_registry_contact",
    "payer_or_financial_access_owner",
}

missing_owners = sorted(required_owners - set(owners))
decision = {
    "gate_label": "mitapivat_owner_routing_ready",
    "owners_checked": len(owners),
    "missing_required_owners": missing_owners,
    "safe_states": sorted(safe_states),
    "policy": "one_question_one_first_owner_one_safe_state",
}

decision
```

## Decision

Continue with owner routing. Outreach must split clinical, pharmacy, regulatory, manufacturer, trial, and financial questions before sending.
