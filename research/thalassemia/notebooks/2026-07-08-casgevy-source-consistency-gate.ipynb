# Experiment: Casgevy source consistency gate

Objective: verify that current explanatory Casgevy notes use the July 2026 FDA 2+ benchmark while dated snapshots remain historical.

```python
from pathlib import Path

repo = Path.cwd()
roots = ["research", "docs", "data", "paper", "scripts", "templates"]
suffixes = {".md", ".json", ".tex", ".bib", ".ipynb"}
terms = ("Casgevy", "CASGEVY", "exagamglogene")
older_age_terms = ("12 years and older", "12 years and over", "age 12+")

hits = []
for root in roots:
    for path in sorted((repo / root).rglob("*")):
        if not path.is_file() or path.suffix not in suffixes:
            continue
        text = path.read_text(encoding="utf-8", errors="ignore")
        if not any(term in text for term in terms):
            continue
        if not any(age_term in text for age_term in older_age_terms):
            continue
        hits.append(str(path.relative_to(repo)))

hits[:12], len(hits)
```

```python
patched_current_notes = {
    "research/thalassemia/findings/2026-04-26-therapy-decision-matrix.md",
    "research/thalassemia/findings/2026-04-27-gene-therapy-access-frontier.md",
    "research/thalassemia/findings/2026-04-27-curative-options-eligibility-gate.md",
    "research/thalassemia/findings/2026-04-28-case001-advanced-therapy-referral-readiness-gate.md",
}
historical_prefixes = (
    "data/literature/",
    "data/workflows/",
    "docs/status/2026-04",
    "docs/status/2026-05",
    "docs/status/2026-06",
)
jurisdiction_specific = {
    "research/thalassemia/references/source-registry.md",
    "paper/references.bib",
    "research/thalassemia/findings/2026-04-26-treatment-landscape.md",
    "research/thalassemia/findings/2026-07-07-casgevy-age-expansion-benchmark.md",
    "research/thalassemia/findings/2026-07-08-casgevy-source-consistency-gate.md",
    "research/thalassemia/notebooks/2026-07-08-casgevy-source-consistency-gate.ipynb",
    "data/regulatory/casgevy/2026-07-07-fda-age-expansion-benchmark.json",
    "data/regulatory/casgevy/2026-07-08-source-consistency-gate.json",
    "docs/status/2026-07-07.md",
    "docs/status/2026-07-08.md",
}

classified = []
for hit in hits:
    if hit in patched_current_notes:
        action = "patched_current_note"
    elif hit in jurisdiction_specific:
        action = "jurisdiction_specific_keep"
    elif hit.startswith(historical_prefixes):
        action = "historical_snapshot_do_not_rewrite"
    else:
        action = "review_if_reader_facing"
    classified.append({"path": hit, "action": action})

unexpected = [row for row in classified if row["action"] == "review_if_reader_facing"]
assert not unexpected, unexpected
classified
```

```python
for path_text in patched_current_notes:
    text = (repo / path_text).read_text(encoding="utf-8")
    assert "STN 125787" in text or "2 years and older" in text, path_text

allowed_outputs = {
    "source_consistency_ready_current_benchmark_only",
    "patched_current_note",
    "historical_snapshot_do_not_rewrite",
    "jurisdiction_specific_keep",
}
blocked_outputs = {
    "eligibility",
    "therapy_selection",
    "trial_screening",
    "referral",
    "travel_import",
    "purchase",
    "dosing",
    "lab_contact",
    "sample_routing",
}
assert not (allowed_outputs & blocked_outputs)

summary = {
    "current_label": "casgevy_source_consistency_ready_current_benchmark_only",
    "hit_count": len(hits),
    "verified_current_note_count": len(patched_current_notes),
    "case001_branch_state": "branch_review_handoff_packet_incomplete",
}
summary
```

Result: current explanatory notes now point to the FDA 2+ benchmark, while older JSON snapshots and dated journals remain historical source records.
