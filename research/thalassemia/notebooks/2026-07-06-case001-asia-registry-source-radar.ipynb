# Experiment: Case-001 Asia registry source radar

Objective: rank public Asia-facing source records for outreach follow-up without producing eligibility, referral, or treatment outputs.

```python
records = [
    {
        "id": "NCT07489196",
        "country": "China",
        "status": "NOT_YET_RECRUITING",
        "kind": "editing",
    },
    {
        "id": "NCT06328764",
        "country": "China",
        "status": "ENROLLING_BY_INVITATION",
        "kind": "editing",
    },
    {
        "id": "NCT06465550",
        "country": "China",
        "status": "RECRUITING",
        "kind": "gene_cell",
    },
    {
        "id": "NCT05477563",
        "country": "Saudi Arabia",
        "status": "RECRUITING",
        "kind": "exa_cel",
    },
    {
        "id": "NCT07292259",
        "country": "Pakistan",
        "status": "COMPLETED",
        "kind": "drug_combo",
    },
    {"id": "PMID41951736", "country": "China", "status": "PUBLISHED", "kind": "paper"},
]

priority = {
    "NOT_YET_RECRUITING": 4,
    "RECRUITING": 3,
    "ENROLLING_BY_INVITATION": 2,
    "PUBLISHED": 2,
    "COMPLETED": 1,
}
ranked = sorted(records, key=lambda row: (-priority[row["status"]], row["id"]))
assert ranked[0]["id"] == "NCT07489196"
ranked
```

```python
blocked_outputs = {
    "eligibility",
    "referral",
    "trial_screening",
    "travel_import",
    "dosing",
    "lab_contact",
    "sample_routing",
}
allowed_output = {
    "label": "asia_registry_source_radar_ready_benchmark_only",
    "top_source": ranked[0]["id"],
    "allowed_use": "source_disambiguation_and_pathway_explanation_only",
}
assert not (set(allowed_output) & blocked_outputs)
allowed_output
```

Result: `NCT07489196` is the highest-priority public follow-up source because it is a China CS-101 Phase 2 benchmark marked not yet recruiting. This is a source-routing result only.
