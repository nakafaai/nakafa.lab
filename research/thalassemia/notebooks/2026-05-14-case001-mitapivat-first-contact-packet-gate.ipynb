# Experiment: Case-001 Mitapivat First Contact Packet Gate

Objective: verify that every first-contact packet has one owner, one question, blocked material, escalation condition, and safe state.

```python
from __future__ import annotations

import json
from pathlib import Path

packet_map = json.loads(
    Path(
        "data/regulatory/mitapivat/2026-05-14-mitapivat-first-contact-packet-map.json"
    ).read_text()
)

required_fields = {
    "owner",
    "first_question",
    "allowed_public_sources",
    "blocked_material",
    "escalate_if",
    "safe_state",
}
missing_fields = {
    packet["owner"]: sorted(required_fields - set(packet))
    for packet in packet_map["packets"]
    if required_fields - set(packet)
}
unsafe_packets = [
    packet["owner"]
    for packet in packet_map["packets"]
    if not packet["blocked_material"] or not packet["escalate_if"]
]

decision = {
    "gate_label": "mitapivat_first_contact_packet_ready",
    "packets_checked": len(packet_map["packets"]),
    "missing_fields": missing_fields,
    "unsafe_packets": unsafe_packets,
    "policy": "one_owner_one_question_public_sources_only",
}

decision
```

## Decision

Continue with first-contact packet discipline. First outreach must stay public-safe and owner-matched before escalation.
