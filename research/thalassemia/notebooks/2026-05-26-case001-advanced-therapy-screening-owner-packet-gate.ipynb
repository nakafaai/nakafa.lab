# Case-001 Advanced-Therapy Screening Owner Packet Gate

This notebook checks that the May 26 owner packet keeps curative biology,
case-specific screening, access, and privacy boundaries separate.

```python
gate = {
    "checked_date": "2026-05-26",
    "gate_label": "case001_advanced_therapy_screening_owner_packet_ready",
    "current_case_label": "advanced_therapy_referral_packet_missing",
    "domains": [
        "diagnosis_genotype",
        "transfusion_burden",
        "immune_blood_bank",
        "iron_organ_chelation",
        "infection_fertility_transplant_readiness",
        "donor_hla_context",
        "access_center_payer",
    ],
    "allowed_public_labels": [
        "advanced_therapy_referral_packet_missing",
        "advanced_therapy_referral_packet_partial",
        "advanced_therapy_referral_packet_ready_for_specialist_screening",
        "under_specialist_review",
        "medically_unsuitable",
        "access_blocked",
        "release_blocked_private_material",
    ],
    "blocked_public_content": [
        "raw_records",
        "screenshots",
        "doctor_messages",
        "clinician_names",
        "hospital_names",
        "owner_replies",
        "contact_details",
        "exact_private_dates",
        "patient_specific_treatment_selection",
        "dose_or_dosing_requests",
        "trial_screening_instruction",
        "travel_or_import_instruction",
    ],
    "source_anchors": [
        "TIF 2025 HSCT chapter",
        "TIF 2025 gene manipulation chapter",
        "FDA CASGEVY product page",
        "FDA ZYNTEGLO product page",
        "May 25 Asia curative gene-cell therapy registry refresh",
        "Quran 16:43 expert consultation",
        "Quran 49:6 verification",
        "Quran 55:7-9 measured claims",
    ],
}

required_domains = {
    "diagnosis_genotype",
    "transfusion_burden",
    "immune_blood_bank",
    "iron_organ_chelation",
    "infection_fertility_transplant_readiness",
    "donor_hla_context",
    "access_center_payer",
}
blocked_jumps = {
    "raw_records",
    "doctor_messages",
    "patient_specific_treatment_selection",
    "trial_screening_instruction",
    "travel_or_import_instruction",
}

errors = []
if set(gate["domains"]) != required_domains:
    errors.append("owner domain set changed")
if gate["current_case_label"] != "advanced_therapy_referral_packet_missing":
    errors.append("case label must remain packet-missing")
if not blocked_jumps.issubset(set(gate["blocked_public_content"])):
    errors.append("blocked public content is incomplete")
if (
    "advanced_therapy_referral_packet_ready_for_specialist_screening"
    not in gate["allowed_public_labels"]
):
    errors.append("ready-for-specialist-screening label is missing")
if len(gate["source_anchors"]) < 8:
    errors.append("source anchors are incomplete")

assert not errors, errors

{
    "gate_label": gate["gate_label"],
    "current_case_label": gate["current_case_label"],
    "domain_count": len(gate["domains"]),
    "blocked_content_count": len(gate["blocked_public_content"]),
}
```
