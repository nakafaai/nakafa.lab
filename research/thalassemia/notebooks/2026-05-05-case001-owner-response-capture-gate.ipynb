# Experiment: Case-001 Owner Response Capture Gate

Objective: map private owner-response types to public-safe labels without raw messages or patient-specific treatment claims.


```python
response_map = {
    "confirm": "captured_label_only",
    "correct": "captured_label_only",
    "request_records": "records_requested",
    "redirect": "owner_redirected",
    "under_review": "owner_under_review",
    "decline": "owner_declined",
    "unclear": "follow_up_needed",
    "private_treatment_instruction": "private_treatment_instruction_omitted",
}

blocked_public_fields = [
    "raw_reply_text",
    "screenshot",
    "clinician_name",
    "hospital_name",
    "phone_number",
    "exact_date",
    "local_path",
    "patient_specific_treatment_instruction",
]

decision = {
    "gate_label": "owner_response_capture_gate_ready",
    "allowed_response_types": sorted(response_map),
    "allowed_public_outputs": sorted(
        set(response_map.values()) | {"capture_pending", "release_blocked"}
    ),
    "blocked_public_fields": blocked_public_fields,
    "public_update_policy": "label_only_after_release_check",
}

decision
```

## Result

The gate allows only label-level public outputs. Any patient-specific treatment instruction is captured privately as present, omitted from public text, and left clinician-owned.
