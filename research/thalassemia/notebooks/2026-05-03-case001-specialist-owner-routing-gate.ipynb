# Experiment: Case-001 Specialist Owner Routing Gate

Objective: assign unresolved case domains to likely clinical owners without using private records or making patient-specific treatment claims.


```python
from __future__ import annotations

from dataclasses import dataclass


@dataclass(frozen=True)
class OwnerRoute:
    """Public-safe clinical owner route for one unresolved case domain."""

    domain: str
    current_label: str
    likely_owner: str
    priority: int
    blocks_patient_relevance: bool
```

```python
routes = [
    OwnerRoute("diagnosis", "phenotype_only", "hematology plus genetics", 5, True),
    OwnerRoute(
        "clinical_dependence",
        "transfusion_dependent_reported",
        "hematology or thalassemia clinic",
        5,
        True,
    ),
    OwnerRoute(
        "transfusion_burden",
        "transfusion_dependent_burden_unquantified",
        "hematology plus transfusion service",
        5,
        True,
    ),
    OwnerRoute(
        "immune_blood_bank",
        "immune_transfusion_packet_missing",
        "transfusion medicine or blood bank",
        4,
        True,
    ),
    OwnerRoute(
        "iron_organ_chelation",
        "iron_packet_missing",
        "hematology plus MRI, cardiology, or endocrine owner",
        4,
        True,
    ),
    OwnerRoute(
        "advanced_therapy_screening",
        "advanced_therapy_referral_packet_missing",
        "thalassemia specialist, transplant, gene-cell therapy, or trial team",
        4,
        True,
    ),
]

owner_states = (
    "owner_confirmed",
    "owner_requested_records",
    "owner_redirected",
    "owner_marked_not_applicable",
    "owner_marked_under_review",
    "owner_unknown",
)

ranked_routes = sorted(routes, key=lambda route: -route.priority)
decision = {
    "gate_label": "specialist_owner_routing_needed",
    "route_order": [(route.domain, route.likely_owner) for route in ranked_routes],
    "allowed_owner_states": list(owner_states),
    "patient_specific_claims": "blocked_until_domain_owner_closes_gate",
}

decision
```

## Result

The highest-priority routes are diagnosis, clinical-dependence category, and transfusion-burden quantification. Every route keeps patient-specific candidate relevance blocked until the responsible clinical owner closes that domain gate.
