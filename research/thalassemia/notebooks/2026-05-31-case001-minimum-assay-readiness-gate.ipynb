# Experiment: Case-001 Minimum Assay Readiness Gate

Objective:
- Verify that May 31 assay readiness requires identity, model, controls, endpoints, stop rules, raw data, and privacy fields.
- Keep assay readiness separate from treatment selection.


```python
# Setup: load the public-safe workflow artifact.
from __future__ import annotations

import json
from pathlib import Path

DATA_PATH = Path("data/workflows/case-001/2026-05-31-minimum-assay-readiness-gate.json")

with DATA_PATH.open() as handle:
    workflow = json.load(handle)

workflow["gate_label"]
```

```python
required_fields = {
    "candidate_identity",
    "candidate_category",
    "target_gap_axis",
    "model_context",
    "positive_control",
    "vehicle_or_negative_control",
    "endpoint_set",
    "safety_stop_rule",
    "replicate_and_raw_data_plan",
    "ethics_and_privacy_boundary",
    "falsification_test",
}

required_endpoints = {
    "hbf_or_hbg_signal",
    "erythroid_maturation",
    "viability_or_apoptosis",
    "hemolysis_or_membrane_damage",
    "globin_chain_balance_or_free_alpha_globin",
    "autophagy_or_alpha_globin_cleanup_when_claimed",
    "access_or_material_feasibility",
}

required_decisions = {
    "assay_ready_for_quote",
    "hold_for_model_or_endpoint_gap",
    "reject_for_identity_or_safety_failure",
    "benchmark_only",
    "ethics_or_privacy_blocked",
}

blocked_content = {
    "patient_samples_without_ethics",
    "patient_specific_treatment_selection",
    "dose_or_dosing_requests",
}

decisions = {item["decision"] for item in workflow["assay_decision_rules"]}

assert workflow["gate_label"] == "case001_minimum_assay_readiness_gate_ready"
assert set(workflow["required_assay_fields"]) == required_fields
assert set(workflow["minimum_endpoint_set"]) == required_endpoints
assert decisions == required_decisions
assert blocked_content <= set(workflow["blocked_public_content"])

summary = {
    "required_field_count": len(workflow["required_assay_fields"]),
    "endpoint_count": len(workflow["minimum_endpoint_set"]),
    "decision_count": len(workflow["assay_decision_rules"]),
}
summary
```

## Results

- The workflow requires eleven assay-readiness fields.
- The workflow requires seven minimum endpoint categories.
- The workflow blocks patient samples, treatment selection, and dosing requests.
