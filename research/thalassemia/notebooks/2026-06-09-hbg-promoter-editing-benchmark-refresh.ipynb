# Experiment: HBG Promoter Editing Benchmark Refresh

Objective: update the novelty gate after 2026 HBG promoter editing evidence.

Done means the notebook returns one deterministic decision label and no private case data.

```python
from __future__ import annotations

benchmarks = [
    ("EDIT-301 / reni-cel Cas12a HBG promoter editing", ["PMID:41931048"], 4, 4),
    (
        "CS-101 base-edited autologous CD34+ cells",
        ["PMID:41951736", "NCT06024876", "NCT06065189", "NCT07489196"],
        4,
        5,
    ),
    ("VGB-Ex01 Cas12b HBG promoter editing", ["NCT06041620"], 2, 3),
]

max_stage = max(item[2] for item in benchmarks)
max_endpoint_overlap = max(item[3] for item in benchmarks)
(max_stage, max_endpoint_overlap)
```

## Gate Logic

A candidate that only claims HbF induction is blocked for novelty because 2026 clinical evidence already covers HBG/HbF restoration. A candidate can stay useful only if it answers a different gap: lower infrastructure, affordability, non-editing validation, and a multi-endpoint assay package.

```python
candidate_lanes = [
    {
        "lane": "HbF-only candidate claim",
        "adds_affordability_gap": False,
        "non_editing_or_low_infrastructure": False,
        "has_dual_readout_plan": False,
        "has_safety_stop": False,
    },
    {
        "lane": "non-editing affordable dual-readout assay candidate",
        "adds_affordability_gap": True,
        "non_editing_or_low_infrastructure": True,
        "has_dual_readout_plan": True,
        "has_safety_stop": True,
    },
]


def classify_lane(lane: dict[str, bool | str]) -> str:
    """Classify a candidate lane against the current benchmark gate."""
    required = [
        "adds_affordability_gap",
        "non_editing_or_low_infrastructure",
        "has_dual_readout_plan",
        "has_safety_stop",
    ]

    if all(bool(lane[field]) for field in required):
        return "assay_gap_candidate"

    return "novelty_blocked_by_hbg_editing_benchmark"


lane_decisions = [
    {"lane": str(lane["lane"]), "decision": classify_lane(lane)}
    for lane in candidate_lanes
]
lane_decisions
```

## Result

The benchmark refresh does not produce a treatment recommendation. It updates the novelty pressure gate for future Nakafa candidate claims.

```python
allowed_gap = (
    "lower-infrastructure non-editing candidate with HbF, "
    "alpha-globin/autophagy, maturation, viability, and hemolysis readouts"
)

summary = {
    "checked_date": "2026-06-09",
    "decision_label": "hbg_promoter_editing_benchmark_refresh_active",
    "max_clinical_stage_score": max_stage,
    "max_endpoint_overlap_score": max_endpoint_overlap,
    "blocked_claim": "HbF rescue or HBG promoter editing is novel by itself",
    "allowed_gap": allowed_gap,
    "lane_decisions": lane_decisions,
}

assert lane_decisions[0]["decision"] == "novelty_blocked_by_hbg_editing_benchmark"
assert lane_decisions[1]["decision"] == "assay_gap_candidate"
summary
```
