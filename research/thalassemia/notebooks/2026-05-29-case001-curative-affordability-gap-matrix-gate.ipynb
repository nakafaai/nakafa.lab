# Experiment: Case-001 Curative Affordability Gap Matrix Gate

Objective:
- Verify that the May 29 workflow names explicit affordability gaps.
- Keep the gate as research strategy, not therapy selection.


```python
# Setup: load the public-safe workflow artifact.
from __future__ import annotations

import json
from pathlib import Path

DATA_PATH = Path(
    "data/workflows/case-001/2026-05-29-curative-affordability-gap-matrix.json"
)

with DATA_PATH.open() as handle:
    workflow = json.load(handle)

workflow["gate_label"]
```

```python
required_axes = {
    "cost",
    "center_infrastructure",
    "myeloablation_or_conditioning_burden",
    "monitoring_and_long_term_follow_up",
    "local_access_and_payer",
    "safety_and_fertility_risk",
    "assay_feasibility",
}

required_routes = {
    "allogeneic_hsct",
    "autologous_gene_cell_therapy",
    "disease_modifying_noncurative_drugs",
    "standard_care_stabilization",
    "registry_trial_benchmark",
}

blocked_content = {
    "patient_specific_treatment_selection",
    "trial_screening_instruction",
    "travel_or_import_instruction",
}

labels = set(workflow["allowed_public_labels"])
routes = {item["route_label"] for item in workflow["route_gap_matrix"]}

assert workflow["gate_label"] == "case001_curative_affordability_gap_matrix_ready"
assert set(workflow["gap_axes"]) == required_axes
assert routes == required_routes
assert blocked_content <= set(workflow["blocked_public_content"])
assert "affordable_cure_research_gap_open" in labels
assert all(
    route["research_use"] != "therapy_recommendation"
    for route in workflow["route_gap_matrix"]
)

summary = {
    "axis_count": len(workflow["gap_axes"]),
    "route_count": len(workflow["route_gap_matrix"]),
    "blocked_content_count": len(workflow["blocked_public_content"]),
}
summary
```

## Results

- The workflow has seven explicit gap axes.
- The workflow keeps all five route branches separate.
- The workflow blocks therapy selection, trial screening, and travel or import instructions.
- Decision: document the gap matrix as a research-strategy gate.
