# Experiment: Case-001 First Quote Panel Assay Readiness Audit

Objective:
- Verify June 1 panel readiness labels and identity refresh fields.
- Keep the first quote panel as preclinical assay feasibility only.


```python
# Setup: load the public-safe workflow artifact.
from __future__ import annotations

import json
from pathlib import Path

DATA_PATH = Path(
    "data/workflows/case-001/2026-06-01-first-quote-panel-assay-readiness-audit.json"
)

with DATA_PATH.open() as handle:
    workflow = json.load(handle)

workflow["gate_label"]
```

```python
required_items = {
    "hydroxyurea",
    "purified resveratrol",
    "sirolimus",
    "standardized 6-shogaol-rich ginger extract",
    "melittin",
    "PF-06409577",
    "T-BDMC-like curcuminoid analog",
}

required_identity_items = {
    "hydroxyurea",
    "resveratrol",
    "sirolimus",
    "6-shogaol",
    "melittin",
    "PF-06409577",
}

blocked_content = {
    "patient_samples_without_ethics",
    "patient_specific_treatment_selection",
    "dose_or_dosing_requests",
}

panel_items = {item["item"] for item in workflow["panel_readiness"]}
identity_items = {item["item"] for item in workflow["identity_refresh"]}
labels = {item["readiness_label"] for item in workflow["panel_readiness"]}

assert workflow["gate_label"] == "case001_first_quote_panel_assay_readiness_audit_ready"
assert panel_items == required_items
assert identity_items == required_identity_items
assert "assay_ready_for_quote" in labels
assert "hold_for_model_or_endpoint_gap" in labels
assert "benchmark_only" in labels
assert blocked_content <= set(workflow["blocked_public_content"])

summary = {
    "panel_count": len(workflow["panel_readiness"]),
    "identity_count": len(workflow["identity_refresh"]),
    "label_count": len(labels),
}
summary
```

## Results

- The workflow audits seven quote-panel items.
- The workflow checks six current PubChem identity records.
- The workflow preserves benchmark-only, quote-ready, and hold labels.
