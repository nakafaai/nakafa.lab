# Case-001 Advanced-Therapy Clinician-Answer Ladder Gate

This notebook checks that clinician answers update only allowed public labels
and do not become therapy, access, travel, import, or trial-screening advice.

```python
owner_domains = {
    "diagnosis_genotype",
    "transfusion_burden",
    "immune_blood_bank",
    "iron_organ_chelation",
    "infection_fertility_transplant_readiness",
    "donor_hla_context",
    "access_center_payer",
}

ready_for_screening = "advanced_therapy_referral_packet_ready_for_specialist_screening"

answer_ladder = {
    "domain_confirmed_record_present": "owner_domain_labeled_present",
    "domain_missing_record_requested": "owner_domain_missing_record_requested",
    "domain_not_applicable_clinician_confirmed": "owner_domain_not_applicable",
    "specialist_screening_reasonable": ready_for_screening,
    "screening_not_reasonable_medical": "medically_unsuitable",
    "access_blocked_owner_confirmed": "access_blocked",
    "private_or_patient_specific_instruction": "release_blocked_private_material",
}

blocked_public_content = {
    "raw_records",
    "screenshots",
    "doctor_messages",
    "clinician_names",
    "hospital_names",
    "owner_replies",
    "contact_details",
    "exact_private_dates",
    "hla_or_donor_details",
    "patient_specific_treatment_selection",
    "dose_or_dosing_requests",
    "trial_screening_instruction",
    "travel_or_import_instruction",
}

required_public_labels = {
    "owner_domain_labeled_present",
    "owner_domain_missing_record_requested",
    "owner_domain_not_applicable",
    "advanced_therapy_referral_packet_ready_for_specialist_screening",
    "medically_unsuitable",
    "access_blocked",
    "release_blocked_private_material",
}

errors = []
if len(owner_domains) != 7:
    errors.append("owner-domain count changed")
if set(answer_ladder.values()) != required_public_labels:
    errors.append("answer ladder does not cover required public labels")
if "hla_or_donor_details" not in blocked_public_content:
    errors.append("HLA or donor details must remain blocked")
if "trial_screening_instruction" not in blocked_public_content:
    errors.append("trial-screening instruction must remain blocked")
if "travel_or_import_instruction" not in blocked_public_content:
    errors.append("travel or import instruction must remain blocked")

assert not errors, errors

{
    "owner_domain_count": len(owner_domains),
    "answer_label_count": len(answer_ladder),
    "blocked_content_count": len(blocked_public_content),
}
```
