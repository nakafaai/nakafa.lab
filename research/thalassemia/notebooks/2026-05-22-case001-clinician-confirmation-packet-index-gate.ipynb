# Experiment: Case-001 Clinician Confirmation Packet Index Gate

Objective:
- Validate the doctor-facing packet order before any disease-specific lane question.
- Confirm the packet blocks raw records, doctor messages, access instructions, and patient-specific treatment claims.

Success criteria:
- seven ordered packet steps are present;
- allowed public response labels are explicit;
- blocked public content is explicit;
- the main paper is appendix-only and mitapivat is downstream of prerequisites.


```python
# Setup: imports and source path
from __future__ import annotations

import json
from pathlib import Path

DATA_PATH = Path(
    "data/workflows/case-001/2026-05-22-clinician-confirmation-packet-index.json"
)
gate = json.loads(DATA_PATH.read_text(encoding="utf-8"))
gate["gate_label"]
```

## Gate Contract

This gate keeps the doctor-facing packet readable. The one-page handoff comes
first, private records stay outside Git, the main paper is appendix-only, and
the mitapivat question is downstream of prerequisite checks.


```python
required_steps = [
    "send_one_page_handoff_first",
    "attach_private_original_records",
    "validate_core_labels",
    "name_missing_records_and_owner",
    "classify_advanced_review_readiness",
    "route_mitapivat_only_after_prerequisites",
    "offer_main_paper_as_appendix_only",
]

required_public_labels = {
    "scope_confirmed_or_corrected",
    "labels_confirmed_corrected_or_unconfirmable",
    "missing_record_owner_labels_only",
    "advanced_review_readiness_label",
    "one_allowed_mitapivat_clinician_response_label",
    "appendix_requested_or_not_needed",
    "release_blocked_private_material",
}

required_blocked_content = {
    "raw_records",
    "exact_private_dates",
    "screenshots",
    "doctor_messages",
    "clinician_names",
    "hospital_names",
    "owner_replies",
    "contact_details",
    "patient_specific_treatment_instructions",
    "dose_or_dosing_requests",
    "access_import_or_travel_instructions",
    "trial_screening_instruction",
}

packet_order = gate["packet_order"]
assert gate["gate_label"] == "case001_clinician_confirmation_packet_index_ready"
assert [item["step_label"] for item in packet_order] == required_steps
assert set(gate["allowed_public_response_labels"]) == required_public_labels
assert set(gate["blocked_public_content"]) == required_blocked_content
assert len(gate["source_checks"]) >= 7

for item in packet_order:
    assert item["deliverable"]
    assert item["doctor_intent"]
    assert item["public_output"]
    assert item["stop_if"]

decision = {
    "gate_label": gate["gate_label"],
    "steps_checked": len(packet_order),
    "public_labels_checked": len(gate["allowed_public_response_labels"]),
    "public_action": "send_handoff_first_keep_main_paper_appendix_only",
}
decision
```

## Results

- The packet starts with the one-page doctor handoff, not the main paper.
- Private source records are attached outside Git and represented publicly as labels only.
- The mitapivat question stays after label validation, missing-record routing, and readiness checks.
- The main paper is explicitly appendix-only.


## Next steps

- Record only public-safe labels after clinician review.
- Return to record retrieval when any required label is wrong, unconfirmable, missing, or release-blocked.
