# Experiment: Case-001 Mitapivat Response Capture Gate

Objective: verify that every owner response label has a public action, private action, and next state before repo capture.

```python
from __future__ import annotations

import json
from pathlib import Path

response_map = json.loads(
    Path(
        "data/regulatory/mitapivat/2026-05-15-mitapivat-response-capture-map.json"
    ).read_text()
)

required_fields = {
    "response_label",
    "meaning",
    "public_repo_action",
    "private_workspace_action",
    "next_state",
}
missing_fields = {
    state["response_label"]: sorted(required_fields - set(state))
    for state in response_map["response_states"]
    if required_fields - set(state)
}
unsafe_states = [
    state["response_label"]
    for state in response_map["response_states"]
    if "raw" in state["public_repo_action"].lower()
]

decision = {
    "gate_label": "mitapivat_response_capture_ready",
    "states_checked": len(response_map["response_states"]),
    "missing_fields": missing_fields,
    "unsafe_public_actions": unsafe_states,
    "policy": "raw_replies_private_public_repo_labels_only",
}

decision
```

## Decision

Continue with response capture. Raw owner replies stay private; public repo updates use labels, owner categories, public sources, and next states only.
