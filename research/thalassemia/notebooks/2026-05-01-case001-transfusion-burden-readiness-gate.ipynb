# Experiment: Case-001 Transfusion Burden Readiness Gate

Objective: decide whether a private de-identified transfusion log is ready for burden calculation without using raw patient data in the public repo.


```python
from __future__ import annotations

from dataclasses import dataclass


@dataclass(frozen=True)
class FieldGate:
    """Readiness requirement for de-identified transfusion burden calculation."""

    name: str
    required_for_first_calc: bool
    explains_short_interval: bool
    available: bool
```

```python
fields = [
    FieldGate("date", True, False, False),
    FieldGate("body_weight_kg", True, False, False),
    FieldGate("transfused_volume_ml", True, False, False),
    FieldGate("red_cell_fraction", True, False, False),
    FieldGate("pre_hb_g_dl", True, False, False),
    FieldGate("post_hb_g_dl", False, True, False),
    FieldGate("product_type", False, True, False),
    FieldGate("reaction_or_delayed_symptom", False, True, False),
    FieldGate("matching_or_antibody_context", False, True, False),
]

missing_core = [
    field.name
    for field in fields
    if field.required_for_first_calc and not field.available
]
missing_explanation = [
    field.name
    for field in fields
    if field.explains_short_interval and not field.available
]

decision = {
    "case_label": "transfusion_dependent_burden_unquantified",
    "readiness_label": "not_ready_missing_core_fields",
    "minimum_rows_required": 2,
    "minimum_window_days": 28,
    "missing_core_fields": missing_core,
    "missing_explanation_fields": missing_explanation,
    "public_case_update_allowed": False,
}

decision
```

## Result

The current public-safe state is not ready for burden interpretation. The next private action is to collect core fields and only then run the calculator for clinician review.
