# Experiment: CS-101 Phase 2 Escalation Gate

Objective: decide how the new `NCT07489196` phase 2 CS-101 registry record changes the Nakafa Lab proximity and novelty gate.

Success criteria:
- load the compact registry snapshot from `data/registries/clinicaltrials/`;
- keep the result framed as benchmark evidence, not treatment advice;
- emit a small decision record that can be copied into the finding and manuscript updates.


```python
from __future__ import annotations

import json
from pathlib import Path

REPO_ROOT = Path.cwd()
SNAPSHOT = (
    REPO_ROOT
    / "data/registries/clinicaltrials"
    / "2026-04-28-nct07489196-cs101-phase2-detail.json"
)

snapshot = json.loads(SNAPSHOT.read_text(encoding="utf-8"))
record = snapshot["records"][0]
record
```

## Plan

Hypothesis: if CS-101 has a phase 2 registry record after the phase 1 Nature report, then weak `increase HbF` novelty claims should be treated as blocked even more strongly.

Metrics:
- phase and status;
- whether the record preserves transfusion-independence, HbF, engraftment, and survival/safety endpoints;
- whether the update changes Nakafa from discovery novelty toward affordability and non-editing validation.


```python
def has_outcome(text: str) -> bool:
    """Return whether the compact primary-outcome text contains a phrase."""
    outcomes = " ".join(record["primary_outcomes"]).lower()
    return text.lower() in outcomes


phase2_gate = {
    "nct_id": record["nct_id"],
    "status": record["status"],
    "phase": record["phases"][0],
    "country": record["countries"][0],
    "intervention": record["interventions"][0],
    "has_safety": has_outcome("adverse events") or has_outcome("saes"),
    "has_survival": has_outcome("overall survival"),
    "has_engraftment": has_outcome("engraftment"),
    "has_transfusion_independence_12m": has_outcome(
        "transfusion independence for at least 12 consecutive months"
    ),
}

phase2_gate
```

## Results

`NCT07489196` does not create a new Nakafa candidate. It upgrades the external CS-101 benchmark from a small phase 1 publication plus registry trail into an active phase 2 development path.

Interpretation: the stronger the CS-101 clinical-development path becomes, the narrower Nakafa Lab's defensible novelty claim becomes. The live gap is affordability, infrastructure, safety, and non-editing validation, not the general idea of HbF rescue through `HBG` promoter/BCL11A biology.


```python
decision = {
    "closest_editing_benchmark": (
        "CS-101 / NCT06024876 plus NCT07489196 phase 2 registry escalation"
    ),
    "novelty_effect": "weak HbF-only claims are novelty-blocked",
    "nakafa_remaining_gap": [
        "affordability_gap",
        "safety_gap",
        "non_editing_validation_gap",
        "integrated_assay_gap",
    ],
    "do_not_claim": "Nakafa has found a cure or that HbF-rescue editing is unexplored",
    "use_as": "external benchmark for lab-pitch and manuscript wording",
}

decision
```
