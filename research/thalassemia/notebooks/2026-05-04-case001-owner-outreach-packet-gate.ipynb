# Experiment: Case-001 Owner Outreach Packet Gate

Objective: convert owner routes into one narrow outreach ask per clinical owner without private records or patient-specific treatment claims.


```python
owner_rows = [
    (
        "hematology_or_thalassemia_clinic",
        5,
        "confirm diagnosis wording, dependence category, and missing records",
    ),
    (
        "transfusion_service_or_blood_bank",
        5,
        "define burden and immune blood-bank record needs",
    ),
    (
        "molecular_genetics_or_genetic_counseling",
        5,
        "define molecular confirmation and modifier-record needs",
    ),
    (
        "iron_mri_cardiology_endocrine_or_organ_monitoring_owner",
        4,
        "define iron and organ-risk packet completeness",
    ),
    (
        "transplant_gene_cell_therapy_or_trial_team",
        3,
        "define screening readiness or blocking data",
    ),
]

public_safe_states = [
    "outreach_not_sent",
    "outreach_sent",
    "records_requested",
    "owner_redirected",
    "owner_declined",
    "owner_under_review",
    "response_received_private",
]

top_priority = max(priority for _, priority, _ in owner_rows)
highest_priority_outreach = [
    owner for owner, priority, _ in owner_rows if priority == top_priority
]

decision = {
    "gate_label": "owner_outreach_packet_ready",
    "current_state": "outreach_not_sent",
    "highest_priority_outreach": highest_priority_outreach,
    "allowed_public_states": public_safe_states,
    "public_updates": "label_only_after_release_check",
    "patient_specific_claims": "blocked_until_owner_response_is_deidentified",
}

decision
```

## Result

The first outreach rows are hematology or thalassemia clinic, transfusion service or blood bank, and molecular genetics or genetic counseling. The main paper remains a background appendix; first contact should use only the relevant owner row.
